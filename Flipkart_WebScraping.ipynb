```python
#importing all the packages
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
base_url = 'https://www.flipkart.com/search?q=macbook&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=macbook%7CLaptops&requestId=725f153c-36bb-478c-8f10-e54c60187a13&as-backfill=on&page='
price_inf = []
prodt_name = []
stars_ = []
specs = []
rating = []
review = []

for i in range(1,3):
  time.sleep(3)
  req = requests.get(base_url+str(i))
  soup = BeautifulSoup(req.text,'html')

  #product name
  product_name = soup.findAll('div',attrs={'class':'_4rR01T'})
  [prodt_name.append(i.text) for i in product_name]

  #price info
  price = soup.findAll('div',attrs={'class':'_30jeq3 _1_WHN1'})
  [price_inf.append(i.text) for i in price]

  #stars 
  stars = soup.findAll('div',attrs={'class':'gUuXy-'})
  [stars_.append(i.span.div.text) for i in stars]

  #ratings and reviews
  ratingreview = soup.findAll('span', attrs = {'class' : '_2_R_DZ'})
  a_l = []
  for i in ratingreview:
    a_l.append(i.span.findAll('span'))
  for i,j,k in a_l:
    rating.append(i.text)
    review.append(k.text)

  #specifications info
  specss = soup.find('ul', attrs = {'class' : '_1xgFaf'})     #specifications  
  specs = specss.findAll('li')
  specs = ','.join([i.text for i in specss])





```

```python
data = pd.DataFrame({'Product':prodt_name,
                     'Price':price_inf,
                     'Stars':stars_,
                     'Ratings':rating,
                     'Reviews':review,
                     'Specifications':specs})
```
